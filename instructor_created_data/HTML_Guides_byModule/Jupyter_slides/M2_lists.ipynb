# Lists

### Creating a List

```python
c = [-45, 6, 0, 72, 1543]
```

```python
c
```

### Accessing Elements of a List
* Reference a list element by writing the list’s name followed by the element’s **index** enclosed in `[]` (the **subscription operator**). 

```python
c[0]
```

```python
c[4]
```

### Determining a List’s Length 

```python
len(c)
```

### Accessing Elements from the End of the List with Negative Indices
* Lists can be accessed from the end by using _negative indices_:


```python
c[-1]
```

```python
c[-5]
```

### Indices Must be either Integers or variables that reference Integers

```python
a = 0
```

```python
b = 3
```

```python
c[a]
```

### Lists Are Mutable (Can be changed)

```python
c[4] = 17
```

```python
c
```

### Appending to a List 

* Using with +=

```python
a_list = []
```

```python
for number in range(1, 6):
    a_list += [number]
```

```python
a_list
```

* Using List methods

```python
a_list.append(9) # adds 9 to the end of the list
```

```python
a_list
```

```python
a_list.extend([45,67]) # adds 45,67 to the end of the list
```

```python
a_list
```
