#  `while` Loop

* Condition tested for true or false value
* Statements repeated as long as condition remains `True`.


```python
score = 3
```

```python
while score < 0:
    print('invalid score entered!!!')  
```

```python
score
```

#### In order for a loop to stop executing, something has to happen inside the loop to make the condition false

* If variable `score` above is assigned a value of -1 , the while loop will iterate (run) infinitely.
* To prevent an infinite loop, something in the `while` suite must change `score`â€™s value, so the condition eventually becomes `False`. 

### `while` Statement Flowchart


```python
score = -1
```

```python
while score < 0:
    print('invalid score entered!!!') 
    score = float (input('Enter scroe again ')) # this statement will prevent an infinite loop from accuring 
```
